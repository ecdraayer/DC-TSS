# SPORTS

```python
time_series = np.loadtxt("./data/Sports_Activity.csv", delimiter=",")
labels = np.loadtxt("./data/Sports_Activity_labels.csv", delimiter=",")
ground_truth = np.where(labels[:-1] != labels[1:])[0]
```

```python
plt.style.use('default')
```

```python
directory = "./1DFloss/SPORTS"
data = []
predictions = []
count = 0
data = np.zeros(170201)
for file in os.listdir(directory):
    if "cac" in file:
        #print(file)
        index = int(re.search(r'\d+', file).group())
        d=np.loadtxt(os.path.join(directory, file))
        data+=d
        count+=1
        #plt.plot(data)
        #print(data.shape)
plt.plot(data/count, color='black')
peaks = find_peaks(-data/count, prominence=0.0001, distance=1000, width=300, height=-0.8)
predictions=peaks[0]
plt.vlines(x = predictions, ymin=0.4, ymax=1,color='red', linestyles='dashed', label="Predicted Change Point",linewidth=1)
plt.vlines(x = ground_truth, ymin=0.4, ymax=1, label="Ground Truth",linewidth=1)
plt.xlabel("Observations Over Time")
plt.ylabel("Combined CAC Value")
plt.title("Combined CAC for Sports Dataset")
plt.legend()

print(peaks[0])
print(len(peaks[0]))
print('covering score:',covering(ground_truth, predictions, len(labels)))
print('f_measure score:',f_measure(ground_truth, predictions, margin=225, alpha=0.5, return_PR=True))
plt.savefig('sportscac.png')
```

```python
predictions
```

```python
reduced_predictions = []
last = np.unique(np.sort(predictions))[0]
true_prediction = last
count = 1
for pred in np.unique(np.sort(predictions))[1:]:
    if pred - last < 500:
        true_prediction += pred
        count+=1
        last = pred 
    else:
        reduced_predictions.append(true_prediction//count)
        true_prediction = pred
        last = pred
        count = 1
reduced_predictions.append(true_prediction//count)
reduced_predictions = np.array(reduced_predictions).astype(int)
print(len(reduced_predictions))
print('covering score:',covering(ground_truth, reduced_predictions, len(labels)))
print('f_measure score:',f_measure(ground_truth, reduced_predictions, margin=225, alpha=0.5, return_PR=True))
```
